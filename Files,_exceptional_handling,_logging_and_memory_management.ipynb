```python
# 2) Write a Python program to read the contents of a file and print each line.
import os

def create_temp_file(filename):
    """Creates a temporary file with sample content."""
    sample_content = (
        "This is the first line.\n"
        "This is the second line, which is slightly longer.\n"
        "And this is the third, final line.\n"
        "One more line for good measure."
    )
    try:
        # Use 'w' mode to create or overwrite the file
        with open(filename, 'w') as f:
            f.write(sample_content)
        print(f"--- Created sample file: '{filename}' ---")
    except IOError as e:
        print(f"Error creating file: {e}")
        return False
    return True

def read_and_print_lines(filename):
    """Reads the contents of the file line by line and prints them."""

    print(f"\n--- Reading contents of '{filename}' line by line ---")

    try:
        # The 'with' statement ensures the file is closed automatically,
        # even if an error occurs.
        with open(filename, 'r') as file:
            # Iterating directly over the file object is the most efficient way
            # to read line by line, as it avoids loading the whole file into memory.
            line_number = 1
            for line in file:
                # The 'line' variable includes the trailing newline character (\n).
                # We use .strip() to remove the newline and any leading/trailing whitespace
                # for cleaner printing, but you can omit it if needed.
                print(f"Line {line_number}: {line.strip()}")
                line_number += 1

    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found.")
    except Exception as e:
        print(f"An unexpected error occurred during file reading: {e}")
    finally:
        # Clean up the temporary file
        if os.path.exists(filename):
            os.remove(filename)
            print(f"\n--- Cleaned up sample file: '{filename}' ---")


if __name__ == "__main__":
    temp_filename = "sample_document.txt"

    if create_temp_file(temp_filename):
        read_and_print_lines(temp_filename)
```

```python
# 4) Write a Python script that reads from one file and writes its content to another file.
import os

# --- Configuration ---
SOURCE_FILE = "source_data.txt"
DESTINATION_FILE = "destination_copy.txt"

def setup_source_file():
    """Creates a sample source file for the copying process."""
    content = (
        "Welcome to the file copying utility.\n"
        "This is the first line from the source.\n"
        "This is the second line.\n"
        "We are testing the read and write operation efficiency."
    )
    try:
        with open(SOURCE_FILE, 'w') as f:
            f.write(content)
        print(f"SETUP: Successfully created source file: '{SOURCE_FILE}'")
    except IOError as e:
        print(f"SETUP ERROR: Could not create source file: {e}")
        return False
    return True

def copy_file_content(source_path, destination_path):
    """
    Reads the content of the source file and writes it to the destination file.

    Uses 'with open' for both files simultaneously for safety and efficiency.
    """
    try:
        # Open both files using a single 'with' statement.
        # Source file is opened for reading ('r').
        # Destination file is opened for writing ('w'). This will create the file
        # or truncate (clear) existing content.
        with open(source_path, 'r') as source, open(destination_path, 'w') as destination:

            # The file object is iterable, so we read line by line.
            # This is efficient for large files as it avoids loading the entire
            # file content into memory at once.

            line_count = 0
            for line in source:
                destination.write(line)
                line_count += 1

        print("\nSUCCESS: File content copied successfully!")
        print(f"   -> Copied {line_count} lines from '{source_path}' to '{destination_path}'.")
        return True

    except FileNotFoundError:
        print(f"\nERROR: Source file '{source_path}' was not found. Cannot proceed with copy.")
    except IOError as e:
        print(f"\nERROR: An I/O error occurred during copy: {e}")
    return False


def verify_copy(destination_path):
    """Reads and prints the content of the destination file to verify the copy."""
    print(f"\n--- Destination File Content ({destination_path}) ---")
    try:
        with open(destination_path, 'r') as f:
            print(f.read().strip())
        print("--------------------------------------------------")
    except Exception as e:
        print(f"Verification Error: {e}")


def cleanup():
    """Removes the generated files."""
    for filename in [SOURCE_FILE, DESTINATION_FILE]:
        if os.path.exists(filename):
            os.remove(filename)
            print(f"CLEANUP: Removed '{filename}'")

# --- Main Execution ---
if __name__ == "__main__":

    # Ensure a clean slate and create the source content
    cleanup()
    if not setup_source_file():
        exit()

    # Perform the copy operation
    if copy_file_content(SOURCE_FILE, DESTINATION_FILE):
        # Verify the results
        verify_copy(DESTINATION_FILE)

    # Final cleanup
    cleanup()
```

```python
# 5) How would you catch and handle division by zero error in Python.

def safe_division(numerator, denominator):
    """
    Performs division and handles ZeroDivisionError.
    """
    try:
        result = numerator / denominator
        print(f"The result of {numerator} / {denominator} is: {result}")
    except ZeroDivisionError:
        print(f"Error: Cannot divide by zero!")
    except TypeError:
        print("Error: Please provide valid numbers for division.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Example usage:
safe_division(10, 2)
safe_division(10, 0)
safe_division(10, "a") # Example of another potential error
```

```python
# 9) How can you read a file line by line and store its content in a list in Python?
import os

def create_sample_file(filename, content):
    """Creates a sample file with the given content."""
    try:
        with open(filename, 'w') as f:
            f.write(content)
        print(f"--- Created sample file: '{filename}' ---")
    except IOError as e:
        print(f"Error creating file: {e}")
        return False
    return True

def read_file_into_list(filename):
    """
    Reads a file line by line and stores each line in a list.
    Removes trailing newline characters.
    """
    lines_list = []
    try:
        with open(filename, 'r') as file:
            for line in file:
                # .strip() removes leading/trailing whitespace, including the newline character
                lines_list.append(line.strip())
        print(f"\nSuccessfully read '{filename}' into a list.")
        return lines_list
    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found.")
        return None
    except Exception as e:
        print(f"An unexpected error occurred during file reading: {e}")
        return None

# --- Example Usage ---
sample_filename = "my_document.txt"
sample_content = (
    "First line of text.\n"
    "Second line.\n"
    "Third and final line."
)

# Create a sample file
if create_sample_file(sample_filename, sample_content):
    # Read the file content into a list
    file_content_list = read_file_into_list(sample_filename)

    # Print the list
    if file_content_list is not None:
        print("Content of the list:")
        print(file_content_list)

    # Clean up the sample file
    if os.path.exists(sample_filename):
        os.remove(sample_filename)
        print(f"\n--- Cleaned up sample file: '{sample_filename}' ---")
```

```python
# 12) Write a program that demonstrates using multiple except blocks to handle different types of exceptions.

def handle_various_errors(value, index):
    """
    Attempts a list operation and handles different potential errors.
    """
    my_list = [1, 2, 3, 4, 5]

    try:
        # Potential errors:
        # 1. TypeError if 'index' is not an integer
        # 2. IndexError if 'index' is out of the list's range
        # 3. ValueError (less likely here, but possible in other list ops)
        # 4. ZeroDivisionError (if 'value' is divided by a calculated result that is zero)
        #    (Adding a division to make ZeroDivisionError possible)
        result = my_list[index] / value

        print(f"Result of accessing list at index {index} and dividing by {value}: {result}")

    except TypeError:
        print(f"Error: Invalid type for index or value. Please provide valid numbers.")
    except IndexError:
        print(f"Error: Index {index} is out of the list's range.")
    except ZeroDivisionError:
        print(f"Error: Division by zero occurred.")
    except Exception as e:
        # A general except block can catch any other unexpected errors
        print(f"An unexpected error occurred: {e}")

# Example usage:
print("--- Testing with valid input ---")
handle_various_errors(2, 3) # Should work (my_list[3] = 4, 4 / 2 = 2.0)

print("\n--- Testing with IndexError ---")
handle_various_errors(1, 10) # Index is out of range

print("\n--- Testing with TypeError for index ---")
handle_various_errors(1, "abc") # Invalid type for index

print("\n--- Testing with TypeError for value ---")
handle_various_errors("xyz", 2) # Invalid type for value

print("\n--- Testing with ZeroDivisionError ---")
handle_various_errors(0.5, 1) # my_list[1] = 2, 2 / 0.5 = 4.0 (Oops, need a case that results in division by zero)

def handle_various_errors_zero_div(value, index):
     """
     Attempts a list operation and handles different potential errors, including ZeroDivisionError.
     """
     my_list = [1, 2, 3, 4, 5]

     try:
         # Potential errors:
         # 1. TypeError if 'index' is not an integer
         # 2. IndexError if 'index' is out of the list's range
         # 3. ZeroDivisionError if 'value' is zero
         result = my_list[index] / value

         print(f"Result of accessing list at index {index} and dividing by {value}: {result}")

     except TypeError:
         print(f"Error: Invalid type for index or value. Please provide valid numbers.")
     except IndexError:
         print(f"Error: Index {index} is out of the list's range.")
     except ZeroDivisionError:
         print(f"Error: Division by zero occurred.")
     except Exception as e:
         # A general except block can catch any other unexpected errors
         print(f"An unexpected error occurred: {e}")

print("\n--- Testing with ZeroDivisionError (Corrected) ---")
handle_various_errors_zero_div(0, 1) # my_list[1] = 2, 2 / 0 = ZeroDivisionError
```

```python
# 15) Write a Python program that prints the content of a file and handles the case when the file is empty.

import os

def create_sample_file(filename, content=""):
    """Creates a sample file with the given content (defaults to empty)."""
    try:
        with open(filename, 'w') as f:
            f.write(content)
        print(f"--- Created sample file: '{filename}' ---")
    except IOError as e:
        print(f"Error creating file: {e}")
        return False
    return True

def print_file_content(filename):
    """
    Reads the content of a file, checks if it's empty, and prints accordingly.
    Handles FileNotFoundError.
    """
    print(f"\n--- Attempting to read '{filename}' ---")
    try:
        with open(filename, 'r') as file:
            content = file.read()

            if content: # Check if the content string is not empty
                print("File content:")
                print(content)
            else:
                print(f"The file '{filename}' is empty.")

    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found.")
    except Exception as e:
        print(f"An unexpected error occurred during file reading: {e}")


# --- Example Usage ---
empty_filename = "empty_file.txt"
non_empty_filename = "non_empty_file.txt"

# Create an empty file
create_sample_file(empty_filename, "")
# Create a non-empty file
create_sample_file(non_empty_filename, "This file has some content.")

# Test with the empty file
print_file_content(empty_filename)

# Test with the non-empty file
print_file_content(non_empty_filename)

# Test with a non-existent file
print_file_content("does_not_exist.txt")

# --- Clean up the sample files ---
if os.path.exists(empty_filename):
    os.remove(empty_filename)
    print(f"\n--- Cleaned up sample file: '{empty_filename}' ---")
if os.path.exists(non_empty_filename):
    os.remove(non_empty_filename)
    print(f"--- Cleaned up sample file: '{non_empty_filename}' ---")
```

```python
# 17) Write a Python program to create and write a list of numbers to a file, one number per line.

import os

def write_numbers_to_file(filename, numbers):
    """
    Writes a list of numbers to a file, one number per line.
    Handles potential IOError during writing.
    """
    print(f"--- Attempting to write numbers to '{filename}' ---")
    try:
        # Open the file in write mode ('w'). This will create the file or overwrite it if it exists.
        with open(filename, 'w') as file:
            for number in numbers:
                # Convert the number to a string and add a newline character
                file.write(str(number) + '\n')

        print(f"Successfully wrote numbers to '{filename}'.")

    except IOError as e:
        print(f"Error writing to file '{filename}': {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# --- Example Usage ---
output_filename = "numbers_list.txt"
my_numbers = [10, 25, 5, 42, 18, 7]

# Write the list of numbers to the file
write_numbers_to_file(output_filename, my_numbers)

# Optional: Verify the content of the file
print(f"\n--- Reading content of '{output_filename}' for verification ---")
try:
    with open(output_filename, 'r') as file:
        print(file.read())
except FileNotFoundError:
    print(f"Error: Verification file '{output_filename}' not found.")
except IOError as e:
    print(f"Error reading verification file: {e}")


# Optional: Clean up the file after execution for demonstration purposes
# if os.path.exists(output_filename):
#     os.remove(output_filename)
#     print(f"\n--- Cleaned up file: '{output_filename}' ---")
```

```python
# 18) How would you implement a basic logging setup that logs to a file with rotation after 1MB ?
import logging
import logging.handlers
import os

# Define the log file name
LOG_FILE = "rotating_log.log"

# Define the maximum size of the log file before rotation (1MB)
MAX_LOG_SIZE = 1 * 1024 * 1024 # 1MB in bytes

# Define the number of backup files to keep
BACKUP_COUNT = 5

# Get a logger instance
logger = logging.getLogger("my_rotating_logger")
logger.setLevel(logging.INFO) # Set the minimum logging level for this logger

# Create a RotatingFileHandler
# filename: The name of the log file
# maxBytes: The maximum size in bytes before rotation
# backupCount: The number of backup files to keep
handler = logging.handlers.RotatingFileHandler(
    LOG_FILE,
    maxBytes=MAX_LOG_SIZE,
    backupCount=BACKUP_COUNT
)

# Create a formatter for the log messages
formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')

# Set the formatter for the handler
handler.setFormatter(formatter)

# Add the handler to the logger
logger.addHandler(handler)

# --- Example Usage ---

print(f"Logging messages to '{LOG_FILE}' with rotation at {MAX_LOG_SIZE} bytes.")
print(f"Keeping up to {BACKUP_COUNT} backup files.")

# Log some messages (you might need to log many messages to trigger rotation)
for i in range(20000): # Adjust the range to produce enough log data
    logger.info(f"This is log message number {i + 1}.")

print("\nLogging complete. Check the directory for log files.")

# Optional: Clean up the files (be careful with this in a real application)
# import glob
# for f in glob.glob(f"{LOG_FILE}*"):
#     os.remove(f)
#     print(f"Removed: {f}")
```

```python
# 19) Write a program that handles both IndexError and KeyError using a try-except block.

def handle_list_and_dict_errors(data, key_or_index):
    """
    Attempts to access data from a list or dictionary and handles IndexError or KeyError.
    """
    try:
        if isinstance(data, list) or isinstance(data, tuple):
            # Attempt to access by index (can raise IndexError)
            value = data[key_or_index]
            print(f"Accessed list/tuple at index {key_or_index}: {value}")
        elif isinstance(data, dict):
            # Attempt to access by key (can raise KeyError)
            value = data[key_or_index]
            print(f"Accessed dictionary with key '{key_or_index}': {value}")
        else:
            print("Error: Input data must be a list, tuple, or dictionary.")

    except IndexError:
        print(f"Error: Index {key_or_index} is out of the range for the sequence.")
    except KeyError:
        print(f"Error: The key '{key_or_index}' was not found in the dictionary.")
    except TypeError:
        print(f"Error: Invalid type used for access (expected integer for list/tuple index or hashable for dictionary key).")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Example Usage:
my_list = [10, 20, 30]
my_dict = {"apple": 1, "banana": 2}

print("--- Testing with a list ---")
handle_list_and_dict_errors(my_list, 1)    # Valid index
handle_list_and_dict_errors(my_list, 5)    # IndexError
handle_list_and_dict_errors(my_list, "a")  # TypeError

print("\n--- Testing with a dictionary ---")
handle_list_and_dict_errors(my_dict, "apple") # Valid key
handle_list_and_dict_errors(my_dict, "cherry") # KeyError
handle_list_and_dict_errors(my_dict, 0)      # KeyError (0 is not a key)

print("\n--- Testing with invalid data type ---")
handle_list_and_dict_errors("hello", 0)    # Valid index for string, but function expects list/tuple/dict
handle_list_and_dict_errors(123, 0)       # Invalid data type
```

```python
# 20) How would you open a file and read its contents using a context manager in Python.

def read_file_with_context_manager(filename):
    """
    Opens a file and reads its contents using a context manager (the 'with' statement).
    Handles FileNotFoundError.
    """
    print(f"--- Attempting to read '{filename}' using a context manager ---")
    try:
        # The 'with' statement ensures the file is automatically closed
        with open(filename, 'r') as file:
            content = file.read()
            print("File content:")
            print(content)
    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found.")
    except IOError as e:
        print(f"An I/O error occurred while reading the file: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Example usage:

# Create a sample file to read from
sample_filename = "sample_for_context.txt"
sample_content = "This is some content in the sample file."
try:
    with open(sample_filename, 'w') as f:
        f.write(sample_content)
    print(f"--- Created sample file: '{sample_filename}' ---")
except IOError as e:
    print(f"Error creating sample file: {e}")

# Read the content using the function with the context manager
read_file_with_context_manager(sample_filename)

# Attempt to read a non-existent file to show error handling
read_file_with_context_manager("non_existent_context_file.txt")

# Optional: Clean up the sample file
import os
if os.path.exists(sample_filename):
    os.remove(sample_filename)
    print(f"\n--- Cleaned up sample file: '{sample_filename}' ---")
```

```python
# 21) Write a Python program that reads a file and prints the number of occurrences of a specific word.

import os
import re # Import the regular expression module for word boundary matching

def create_sample_file(filename, content):
    """Creates a sample file with the given content."""
    try:
        with open(filename, 'w') as f:
            f.write(content)
        print(f"--- Created sample file: '{filename}' ---")
    except IOError as e:
        print(f"Error creating file: {e}")
        return False
    return True

def count_word_occurrences(filename, word_to_count):
    """
    Reads a file and counts the occurrences of a specific word.
    Handles FileNotFoundError and potential reading errors.
    """
    count = 0
    print(f"\n--- Attempting to count occurrences of '{word_to_count}' in '{filename}' ---")
    try:
        with open(filename, 'r') as file:
            content = file.read()

            # Use regular expressions to find the word, ensuring it's a whole word
            # re.findall finds all non-overlapping matches
            # re.escape escapes special characters in the word
            # \b matches word boundaries
            word_pattern = re.compile(r'\b' + re.escape(word_to_count) + r'\b', re.IGNORECASE) # Case-insensitive count
            matches = word_pattern.findall(content)
            count = len(matches)

        print(f"The word '{word_to_count}' appears {count} times in '{filename}'.")
        return count

    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found.")
        return -1 # Indicate an error
    except IOError as e:
        print(f"An I/O error occurred while reading the file: {e}")
        return -1 # Indicate an error
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return -1 # Indicate an error

# --- Example Usage ---
sample_filename = "sample_text.txt"
sample_content = (
    "This is a sample text file. It contains the word sample multiple times.\n"
    "Sample is a common word. Another sample line.\n"
    "Let's see how many times 'sample' (case-insensitive) appears.\n"
    "SAMple, sample, SAMPLE."
)
word_to_find = "sample"

# Create a sample file
if create_sample_file(sample_filename, sample_content):
    # Count the occurrences of the word
    count_word_occurrences(sample_filename, word_to_find)

    # Test with a non-existent file
    count_word_occurrences("non_existent_text_file.txt", "word")

    # Clean up the sample file
    if os.path.exists(sample_filename):
        os.remove(sample_filename)
        print(f"\n--- Cleaned up sample file: '{sample_filename}' ---")
```

```python
# 22) How can you check if a file is empty before attempting to read its contents.

import os

def is_file_empty(filepath):
    """
    Checks if a file exists and if it is empty.

    Returns True if the file exists and is empty, False otherwise.
    Prints messages indicating the result.
    """
    if not os.path.exists(filepath):
        print(f"Info: File '{filepath}' does not exist.")
        return False  # File doesn't exist, so it's not empty (in the sense of having content)
    else:
        try:
            file_size = os.path.getsize(filepath)
            if file_size == 0:
                print(f"The file '{filepath}' exists and is empty (0 bytes).")
                return True
            else:
                print(f"The file '{filepath}' exists and is not empty ({file_size} bytes).")
                return False
        except OSError as e:
            print(f"Error accessing file '{filepath}': {e}")
            return False # Return False in case of an error accessing the file


# --- Example Usage ---
empty_filename = "test_empty_file.txt"
non_empty_filename = "test_non_empty_file.txt"
non_existent_filename = "test_non_existent_file.txt"

# Create a temporary empty file
try:
    with open(empty_filename, 'w') as f:
        pass # Create an empty file
    print(f"\n--- Created empty file: '{empty_filename}' ---")
except IOError as e:
    print(f"Error creating empty file: {e}")


# Create a temporary non-empty file
try:
    with open(non_empty_filename, 'w') as f:
        f.write("This file has some content.")
    print(f"--- Created non-empty file: '{non_empty_filename}' ---")
except IOError as e:
    print(f"Error creating non-empty file: {e}")


# Check the files
print("\n--- Checking file emptiness ---")
is_file_empty(empty_filename)
is_file_empty(non_empty_filename)
is_file_empty(non_existent_filename)

# Clean up the temporary files
print("\n--- Cleaning up files ---")
if os.path.exists(empty_filename):
    os.remove(empty_filename)
    print(f"Removed: '{empty_filename}'")
if os.path.exists(non_empty_filename):
    os.remove(non_empty_filename)
    print(f"Removed: '{non_empty_filename}'")
```

```python
# 23) Write a Python program that writes to a log file when an error occurs during file handling.

import logging
import os

# Define the log file name
LOG_FILE = "file_handling_errors.log"

# Configure logging to write error messages to the log file
logging.basicConfig(
    filename=LOG_FILE,
    level=logging.ERROR,  # Log only messages with level ERROR or higher
    format='%(asctime)s - %(levelname)s - %(message)s'
)

def write_to_file_safely(filename, content):
    """
    Attempts to write content to a file and logs an error if it fails.
    """
    print(f"--- Attempting to write to '{filename}' ---")
    try:
        # Attempt to open the file for writing ('w')
        with open(filename, 'w') as file:
            file.write(content)
        print(f"Successfully wrote to '{filename}'.")

    except IOError as e:
        error_message = f"An IOError occurred while writing to '{filename}': {e}"
        logging.error(error_message)  # Log the error message
        print(f"Error writing to file: {e}. Details logged to {LOG_FILE}")
    except Exception as e:
        error_message = f"An unexpected error occurred while writing to '{filename}': {e}"
        logging.error(error_message)
        print(f"An unexpected error occurred: {e}. Details logged to {LOG_FILE}")


# --- Example Usage ---
output_filename = "my_output_file.txt"
data_to_write = "This is some data to write to the file."

# Case 1: Successful file writing
write_to_file_safely(output_filename, data_to_write)

# Case 2: Simulate a file handling error (e.g., trying to write to a protected location)
# This might not cause an error in a Colab environment unless you try to write to a
# restricted system directory. We'll demonstrate with a path that is likely to cause
# an issue or you can modify it to a path you know will fail.
# Example: Try to write to a directory you don't have permissions for (might not work on all systems)
# error_filename = "/root/restricted_file.txt" # Example of a potentially restricted path
# write_to_file_safely(error_filename, "This should fail.")

# A more reliable way to simulate an error for demonstration:
# Close the file handle prematurely (not typical in real code, but shows logging)
print("\n--- Simulating a file writing error ---")
try:
    # Open file normally
    file_handle = open("simulated_error.txt", "w")
    # Artificially close it immediately
    file_handle.close()
    # Now try to write using the closed handle
    file_handle.write("This should cause an error.")
except ValueError as e:
    error_message = f"A ValueError occurred during simulated file writing: {e}"
    logging.error(error_message)
    print(f"Simulated error caught and logged: {e}")
except Exception as e:
    error_message = f"An unexpected error occurred during simulated file writing: {e}"
    logging.error(error_message)
    print(f"An unexpected error occurred during simulated file writing: {e}. Details logged to {LOG_FILE}")
finally:
    # Ensure the simulated file is cleaned up if created
    if os.path.exists("simulated_error.txt"):
        os.remove("simulated_error.txt")

print(f"\nCheck the file '{LOG_FILE}' for error logs.")

# Optional: Clean up the log file
# if os.path.exists(LOG_FILE):
#     os.remove(LOG_FILE)
#     print(f"Cleaned up log file: '{LOG_FILE}'")
```
